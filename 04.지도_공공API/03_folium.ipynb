```python
map=folium.Map(location=[37.5598688323642,126.967109652664],zoom_start=14)
folium.Marker(location=[37.5598688323642,126.967109652664],popup='한국경제신문사').add_to(map)
folium.Marker(location=[37.5663174209601,126.977829174031],tooltip='서울특별시청',icon=folium.Icon(color='red',icon='info-sign')).add_to(map)
folium.Circle(
    radius=100,
    location=[37.5641201543296,126.998009728978],
    popup="중구청",
    color="crimson",
    fill=False,
).add_to(map)
map
```

- addr2.csv 를 읽어서 지도위에 표시하기

```python
df = pd.read_csv('addr2.csv')
df
```

```python
map = folium.Map(
    location=[df['위도'].mean(), df['경도'].mean()], 
    zoom_start=13
)
for i in df.index:
    folium.Marker(
        location=[df['위도'][i], df['경도'][i]],
        tooltip=df['건물명'][i],
        popup=folium.Popup(
            df['주소'][i], min_width=200, max_width=200)
    ).add_to(map)
title_html = '<h3 align="center" style="font-size:20px"><b>5개 지역 표시</b></h3>'   
map.get_root().html.add_child(folium.Element(title_html))
map
```
