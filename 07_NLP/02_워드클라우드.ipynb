```python
import konlpy
konlpy.__version__
```

```python
# Open korean text
from konlpy.tag import Okt
okt = Okt()
```

```python
okt.nouns('해당 에러가 발생하는 것을 알 수 있다')
```

```python
with open('data/gift.txt','r',encoding='utf-8') as f:
    gift_text = f.read()
```

```python
tokens= okt.nouns(gift_text)
tokens[:10]
```

```python
# 영문자, 숫자 제거 
import re 
new_tokens = []
for token in tokens:
    text = re.sub('[A-Za-z0-9]','',token)
    if text:
        new_tokens.append(text)
```

```python
import numpy as np 
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib as mpl
%matplotlib inline
plt.rcParams["font.family"] = 'Malgun Gothic'
mpl.rcParams['axes.unicode_minus'] =False
```

```python
import nltk
```

- 아래 두 셀을 반복해서 실행하여 stop_words를 갱신한다. 

```python
stop_words= ['것']
new_tokens=[word for word in new_tokens if word not in stop_words]
new_tokens[:10]
```

```python
gf_gift = nltk.Text(new_tokens, name = '여친선물')
plt.figure(figsize=(15,6))
gf_gift.plot(50)
plt.show()
```

### 워드 클라우드 그리기

```python
import numpy as np 
from PIL import Image
from wordcloud import WordCloud
```

```python
data = gf_gift.vocab().most_common(300)
wc = WordCloud(
    font_path='C:\Windows\Fonts\malgun.ttf',
    relative_scaling=0.2,
    background_color='white'
).generate_from_frequencies(dict(data))
```

```python
plt.figure(figsize=(12,6))
plt.imshow(wc)
plt.axis('off')
plt.show()
```
