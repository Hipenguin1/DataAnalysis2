- Axes 객체와 subplot

```python
x1 = np.linspace(0., 5.) #default 50
x2 = np.linspace(0., 2.) 
y1 = np.cos(2 * np.pi * x1) * np.exp(-x1)
y2 = np.cos(2 * np.pi * x2)
```

```python
ax1 = plt.subplot(2,1,1)
plt.title('2 Subplots')
ax1.plot(x1,y1,'yo-')
ax1.set_ylabel('Damped oscillation')

ax2 = plt.subplot(2,1,2)
ax2.plot(x2,y2,'r.-')
ax2.set_ylabel('Undamped')

plt.xlabel('time (sec')
plt.show()  
```

```python
ax1 = plt.subplot(211)
plt.title('2 Subplots')
plt.plot(x1,y1,'yo-')
plt.ylabel('Damped oscillation')

ax2 = plt.subplot(212)
plt.plot(x2,y2,'r.-')
plt.ylabel('Undamped')
plt.xlabel('time (sec')

plt.tight_layout()
plt.grid(); ax2.grid()
plt.show()  
```

```python
np.random.seed(0)

plt.subplot(221)
plt.plot(np.random.rand(5))
plt.title("axes 1")

plt.subplot(222)
plt.plot(np.random.rand(5))
plt.title("axes 2")

plt.subplot(223)
plt.plot(np.random.rand(5))
plt.title("axes 3")

plt.subplot(224)
plt.plot(np.random.rand(5))
plt.title("axes 4")

plt.tight_layout()
plt.show()
```

```python
fig, axes = plt.subplots(2,2)

np.random.seed(0)
axes[0, 0].plot(np.random.rand(5))
axes[0, 0].set_title("axes 1")
axes[0, 1].plot(np.random.rand(5))
axes[0, 1].set_title("axes 2")
axes[1, 0].plot(np.random.rand(5))
axes[1, 0].set_title("axes 3")
axes[1, 1].plot(np.random.rand(5))
axes[1, 1].set_title("axes 4")

plt.tight_layout()
plt.show()
```

```python
fig, axes = plt.subplots(2,2)

np.random.seed(0)
for i in range(4):
    ax = axes[i//2, i%2]
    ax.plot(np.random.rand(5))
    ax.set_title(f"axes {i+1}")

plt.tight_layout()
plt.show()
```
