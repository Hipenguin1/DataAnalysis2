```python
import numpy as np 
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib as mpl
import seaborn as sns

%matplotlib inline
```

```python
plt.rcParams["font.family"] = 'Malgun Gothic'
mpl.rcParams['axes.unicode_minus'] =False
```

```python
iris = sns.load_dataset('iris')
titanic = sns.load_dataset('titanic')
tips = sns.load_dataset('tips')
flights = sns.load_dataset('flights')
```

```python
sns.countplot(x='class',data=titanic)
plt.title('타이타닉호의 각 클래스별 승객 수 ')
plt.show()
```

```python
sns.countplot(x='day',data=tips)
plt.title('요일별 팁을 준 횟수')
plt.show()
```

- pairplot()

```python
sns.pairplot(iris)
plt.title('Iris 데이터의 Pair Plot')
plt.show()
```

```python
#차원축소 
pca 주성분 분석 
```

```python
flights_passengers = flights.pivot("month", "year", "passengers")
plt.figure(figsize=(8,6))
plt.title("연도, 월 별 승객수에 대한 Heatmap")
sns.heatmap(flights_passengers, annot=True, fmt="d", linewidths=1)
plt.show()
```

```python
sns.barplot(x="day", y="total_bill", data=tips)
plt.title("요일 별, 전체 팁")
plt.show()
```

```python
sns.barplot(x="day", y="total_bill", hue='sex', data=tips)
plt.title("요일 별/성별 전체 요금의 평균")
plt.show()
```

- boxplot 

```python
sns.boxplot(x="day", y="total_bill", data=tips)
plt.title("요일 별, 전체 요금의 Box Plot")
plt.show()
```
